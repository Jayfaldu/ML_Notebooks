```
def shuffle(df,field):
  df=df.copy()
  z=df[field].sample(n=len(df),random_state=1)
  z.reset_index(inplace=True,drop=True)
  df[field]=z
  return df
```

```
def feature_importance(m,x):
  score=m.predict(x)
  fi={}
  for t in x.columns:
    x_shuff=shuffle(x,t)
    fi[t]=((m.predict(x_shuff) - score)**2).mean()
  return fi  
```
